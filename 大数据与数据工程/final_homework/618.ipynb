读取txt文件

```python
with open('./618.txt',mode = 'r',encoding='utf-8') as f:
    text = f.read() #返回的是字符串
text  #所有内容组成的字符串
```

```python
#正则表达式
re_pat = r'[\d-]{10}\s[\d:]{7,8}\s.*[\)>]'
```

```python
log_title_list = re.findall(re_pat,text) 
log_title_list
```

```python
log_content_list = re.split(re_pat,text) #记录内容,re_pat做分隔符，将文本进行分割
log_content_list
```

```python
print(len(log_title_list))
print(len(log_content_list))
```

两个数据量并不一致，因为内容导出时，会有最前面的一行消息记录故需要做删除处理

```python
#头与内容数量应保持一致，现发现内容多1，是因为多了第1行
log_content_list.pop(0) #删除第一行
len(log_content_list)
```

```python
#将记录头及记录内容存储到dataframe中
df = pd.DataFrame()
df['title'] = log_title_list
df['content'] = log_content_list
df
```

```python
#分离日期、用户名、QQ号
df['time'] = df['title'].str.findall('\d{4}-\d{2}-\d{2}\s\d{1,2}:\d{2}:\d{2}').str.get(0)
df['name'] = df['title'].str.findall('\d{1,2}:\d{2}:\d{2}(.*)[\(<]').str.get(0) #提取用户名 从时间开始一直到'('为止，要的数据用'()'括起来
df['QQ'] =  df['title'].str.findall('[\(<]([\dA-z\.@]*)[\)>]').str.get(0) #以'('或'<'开头，以')'或'>'结尾，中间部分可以是数字、字母、.、@的组合
```

```python
df
```

```python
#删除空格与换行符
df['content'] = df['content'].str.replace('\n','')
df
```

```python
#title列内容已被拆分，可以进行删除
df.drop(['title'],axis = 1,inplace = True)
df
```

```python
# 删除系统提示消息
df = df[~df['QQ'].isin(['1000000','10000'])] 
df.reset_index(drop = True,inplace = True)
df
```

```python
df = df[df['content'] != '[图片]']
df = df[df['content'] != '[表情]']
df
```

```python
#内容拼接起来
content_txt = df['content'].str.cat(sep = '。')
#加载停用词
with open('cn_stopwords.txt',encoding = 'utf-8') as file:
    stopword = file.read()
import jieba
words = jieba.lcut(content_txt)
words = [word for word in words if (word not in stopword.split('\n'))&(len(word)>1)]
#大致看看是否已将无用信息去除完毕，若没有还需进行清洗
word_s = pd.Series(words).value_counts()
word_s
```

```python

words = jieba.lcut(content_txt)

# 过滤停用词和长度小于等于1的词语
words = [word for word in words if (word not in stopword.split('\n')) and len(word) > 1]

# 定义一个函数用来判断是否全为数字
def contains_digits(word):
    return any(char.isdigit() for char in word)

# 过滤包含数字的词语
words = [word for word in words if not contains_digits(word)]

# 将列表转换为 Series 并统计词频
word_s = pd.Series(words).value_counts()

# 打印词频统计结果
word_s
```

```python
university_alias = {
    '清华大学': ['thu','th', '清华', 't大','贵系','茶园','叉院','清软'],
    '北京大学': ['pku','pk' '北大','p大','rw','软微','北叉','北计'],
    '复旦大学': ['fdu','fd' '复旦', '蛋蛋','旦大','旦旦','邯郸路'],
    '上海交通大学': ['sjtu', 'sj','上交','交大'],
    '浙江大学': ['zju', '浙大','浙软','紫金港'],
    '中国科学技术大学': ['ustc', '中科大','科大'],
    '南京大学': ['nju', '南大','南软'],
    '同济大学': ['同济', 'tongji', 'tongj','黄渡理工','嘉定','tju']
}
```

```python
def extract_university(context):
    for official_name, alias_list in university_alias.items():
        for alias in alias_list:
            if alias in context:
                return official_name
    return None
extract_university('黄渡理工')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Your data processing code
df_new = df

df_new['content'] = df_new['content'].str.lower()
df_new['university'] = df_new['content'].apply(extract_university)
df_new = df_new[df_new['university'].notnull()]

# 提取时间中的小时
df_new['hour'] = df_new['time'].dt.hour

df_new = df_new.groupby(['hour', 'university']).size().reset_index(name='counts')
df_new.sort_values(by=['hour', 'counts'], ascending=[True, False], inplace=True)

# 找到每个大学聊天量的峰值
max_counts = df_new.groupby('university')['counts'].idxmax()
peak_hours = df_new.loc[max_counts]

plt.figure(figsize=(15,10))
sns.lineplot(data=df_new, x='hour', y='counts', hue='university', linewidth=2.5)

# Set xticks every hour
plt.xticks(np.arange(24), [str(i) + 'h' for i in range(24)])

# 在峰值处标注
for _, row in peak_hours.iterrows():
    plt.text(row['hour'], row['counts'], int(row['counts']), verticalalignment='bottom')

plt.show()


```
